# Digit Recognition Notebook

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
train_dataset = pd.read_csv('datasets/train.csv')
```

```python
train_dataset.head()
```

```python
num_pixels = train_dataset.shape[1] - 1
num_labels = len(train_dataset.label.unique())
ys = tf.one_hot(train_dataset.label, depth=num_labels)
xs = train_dataset.iloc[:, 1:]
xs.head()
```

### *Creating a model*

```python
# Define Sequential model with 3 layers
model = keras.Sequential(
    [
        keras.Input(shape=(num_pixels,)),
        layers.Dense(50, activation="sigmoid", name="hidden_layer1"),
        layers.Dense(20, activation="sigmoid", name="hidden_layer2"),
        layers.Dense(10, activation="sigmoid", name="output_layer"),
    ]
)

model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
class StopOnPoint(tf.keras.callbacks.Callback):
    def __init__(self, point):
        super(StopOnPoint, self).__init__()
        self.point = point

    def on_epoch_end(self, epoch, logs=None): 
        accuracy = logs["accuracy"]
        self.epoch_end=epoch+1
        if accuracy >= self.point:
            self.model.stop_training = True
            
```

```python
stop = StopOnPoint(0.95)
model.fit(
    x=xs,
    y=ys,
    epochs=10,
    validation_split=0.3,
    callbacks=[stop]
)
print(f'Training finished after {stop.epoch_end} epochs')
```

```python
test_dataset = pd.read_csv('datasets/test.csv')
test_dataset.head()
```

```python
outfile = 'submission.csv'

res = model.predict(test_dataset)
results = []
for i, res in enumerate(res):
    r = {'ImageId': i + 1, 'Label': np.argmax(res)}
    results.append(r)
    
submission = pd.DataFrame(results, columns=['ImageId', 'Label'])
submission.head()
```

```python
submission.to_csv(outfile, index=False)
```
